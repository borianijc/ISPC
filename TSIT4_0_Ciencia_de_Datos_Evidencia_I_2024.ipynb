#Qu es Pandas profiling?
Pandas Profiling es una herramienta de anlisis exploratorio de datos (EDA) en Python que ofrece una forma rpida y sencilla de obtener informacin detallada sobre un conjunto de datos.

Funciona en tres pasos principales:

 * Importacin de datos: Puedes importar datos desde diversas fuentes, como archivos CSV, JSON, Excel o Pandas DataFrames.
 * Generacin de informes: Analiza automticamente los datos y genera un informe completo que incluye estadsticas descriptivas, distribuciones, correlaciones y ms para cada variable en el conjunto de datos.
 * Visualizacin de resultados: Ofrece visualizaciones interactivas que ayudan a comprender mejor la estructura y las relaciones dentro de los datos, incluyendo histogramas, diagramas de dispersin, matrices de correlacin y ms.

#Caractersticas claves de Pandas Profiling:

* Facilidad de uso: No se necesita experiencia en programacin para utilizarlo, ya que automatiza gran parte del proceso de EDA.
* Amplia gama de informacin: Proporciona una variedad de estadsticas descriptivas y visualizaciones para cada variable, lo que permite una comprensin exhaustiva de los datos.
* Personalizacin: Aunque es automatizado, permite cierto grado de personalizacin para adaptarse a las necesidades especficas del usuario.
* Comparacin de conjuntos de datos: Permite comparar dos conjuntos de datos para identificar diferencias significativas en trminos de distribuciones, correlaciones y ms.


En resumen, Pandas Profiling es una herramienta poderosa para realizar anlisis exploratorio de datos en Python, proporcionando informacin detallada y visualizaciones interactivas para comprender rpidamente la estructura y las relaciones dentro de los datos.

#Leemos el archivo para analizar

* En este caso, una base de datos de clientes de un servicio de telecomunicaciones, donde podemos ver el estado aquellos que permancen como clientes y los que se fueron, de estos, podemos ver los motivos y/o la razn por la cual solicitaron la baja del servicio.
* Vemos que el DataFrame contiene 7043 registros o filas por 8 columnas de datos.


**Se muestra el Data Frame con la cantidad de filas y columnas.**

```python
analisis_servicios2 = pd.read_csv("Telco_customer_churn_status.csv")
analisis_servicios2
```

**Con el comando .info() de pandas podemos ver las variables, la cantidad de datos no nulos que tienen y el tipo de datos que contienen.**
* Podemos ver como las variables 'Churn Category' y 'Churn Reason' tiene solo 1869 datos correctos, que son los registros de los clientes que dieron de baja el servicio. Los cuales a continuacin haremos una extraccion de esos datos para realizar un analisis de los motivos y/o razones.

```python
analisis_servicios2.info()
```

**Con el comando .isnull anidando la funcin sum() podremos saber la cantidad de datos nulos que tiene la BBDD por variable.**
* Siguiendo lo mencionado anteiormente, comprobamos que las variables 'Churn Category' y 'Churn Reason' tiene 5174 datos nulos, que son los registros de los clientes que no dieron de baja el servicio.

```python
analisis_servicios2.isnull().sum().to_frame()
```

**A continuacin con la funcin .dropna() y en este caso, seleccionando las variables de inters como 'Churn Category' y 'Churn Reason', eliminaremos los registros que sean nulos.**
* Luego de ejecutar podremos ver como ahora el DataFrame nos queda de 1869 filas o registros x las mismas 8 columnas que tenamos.

```python
analisis_servicios2 = analisis_servicios2.dropna(subset=['Churn Category', 'Churn Reason'])
analisis_servicios2
```

**Teniendo en cuanta las variables, como podremos ver en el informe propuesto por la librera pandas profiling, identificamos variables de las cuales podemos decir:**
* **Customer ID:** Es unique, ya que para cada cliente va a haber un nico Id.
* **Count:** Como una constante en este caso, pero podra referirse a la cantidad de cuentas por usuario.
* **Quarter:** Es una variable categrica, se refiere al trimestre, en este caso todo el set es del Q3, pero si ampliamos la BBDD, podremos distinguir por trimestre del ao.
* **Customer Status:** Variable categrica, que nos permite ver el estado del cliente, luego de la limpieza nos quedamos solo con aquellos que desconectaron el servicio.
* **Churn Label:** Es un dato Booleano donde indica si contiene datos del registro que dio de baja el servicio en formato (yes/no)  
* **Churn Value:** Mismo caso que el anterior, indica 0 si no esta desconectado y 1 si es de tipo Churned o desconectado.
* **Churn Category:** Variable categrica que nos indica la categora de desconexin
* **Churn Reason:** Variable categrica donde podremos conocer las razones de desconexin.

**Utilizando Pandas, con la sigueinte sentencia vemos que la cantidad de registros corresponden a los clientes que dieron de baja el servicio.**

```python
customer_status = analisis_servicios2['Customer Status'].value_counts().to_frame()
customer_status
```

**Adems, con Pandas podemos ver las categoras de las desconecciones y las razones de desconexion y la cantidad de registros por las distintas categoras y razones.**

```python
churned_category = analisis_servicios2['Churn Category'].value_counts().to_frame()
churned_category
```

```python
churned_reason = analisis_servicios2['Churn Reason'].value_counts().to_frame()
churned_reason
```


**Utilizando la Pandas Profiling, con el comando ProfileReport () nos muestra un reporte con todas las variables del DF para poder ver las posibilidades de analisis.**
